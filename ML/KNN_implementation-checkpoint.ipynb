In the cell below, develop your own code for performing k-Nearest Neighbour classification.  You may use the scikit-learn k-NN implementation from the labs as a guide - and as a way of verifying your results - but it is important that your implementation does not use any libraries other than the basic numpy and matplotlib functions. 

Define a function that performs k-NN given a set of data.  Your function should be invoked similary to:

        y_ = mykNN(X,y,X_,options)
        
where X is your training data, y is your training outputs, X\_ are your testing data and y\_ are your predicted outputs for X\_.  The options argument (can be a list or a set of separate arguments depending on how you choose to implement the function) should at least contain the number of neighbours to consider as well as the distance function employed.



```python
# helper code
def train_test_split(X, y, test_size=0.25):
    """
    This function splits a dataset into training and test sets

    Parameters:
    ----------
    X: numpy array containing the input data
    y: numpy array containing the target labels
    test_size: float value between 0 and 1, it represents the proportion of the data that should be in the test set. Default value is 0.25.
 
    """
    
    # Generate a random array of indices
    indices = np.random.permutation(len(X))
    
    # Calculate the size of the test set
    test_set_size = int(len(X) * test_size)
    
    # Split the indices into test and train sets
    test_indices = indices[:test_set_size]
    train_indices = indices[test_set_size:]
    
    # Split the data into test and train sets
    X_test, y_test = X[test_indices], y[test_indices]
    X_train, y_train = X[train_indices], y[train_indices]
    
    # Return the train and test sets
    return X_train, X_test, y_train, y_test

def calc_euclidean(v1, v2):
    return np.sqrt(np.sum((v1 - v2)**2))

def calc_manhattan(v1, v2):
    return np.sum(abs(v1 - v2))
```

```python
# mykNN code
def get_knn(x_train, y_train, x_test, k, distance_metric='euclidean'):
    """
    This function splits a dataset into training and test sets

    Parameters:
    ----------
    x_train: a numpy array containing the training data
    y_train: a numpy array containing the training labels
    x_test: a numpy array containing the test data
    k: an integer representing the number of nearest neighbors to consider
    distance_metric: a string representing the distance metric to use. The default value is 'euclidean', which means that the Euclidean distance will be used. Other possible value is 'manhattan', which means that the Manhattan distance will be used.
 
    """
    
    # Initialize an empty list to store the predictions
    predictions = []
    # Loop through the test data points
    for i in x_test:
        stored_dist = []
        # Calculate the distance between each test point and each training data point, store it in a list
        for j in range(len(x_train)):
            
            if distance_metric == 'euclidean':
                distance = calc_euclidean(x_train[j,:], i)
            elif distance_metric == 'manhattan':
                distance = calc_manhattan(x_train[j,:], i)
                
            stored_dist.append(distance)
            
        # Sort the distances in ascending order and filter to the k-nearest neighbors based on their indices.
        sorted_dist = np.argsort(stored_dist)[:k]

        # filter the training target values to the the k-nearest neighbors
        label = y_train[sorted_dist]
        # find the most common label : this is the predicted label
        lab = mode(label)[0][0] 
        # apprend the predicted label to the list of predictions
        predictions.append(lab)
    
    return predictions
```

In the cell below, implement your own classifier evaluation code. This should include some way of calculating confusion matrices, as well as common metrics like accuracy.

```python
# confusion matrix, accuracy, precision, recall, etc. 
def calc_confusion_matrix(y_true, y_pred):
    """
    Returns a confusion matrix, a table that shows the number of correct and incorrect predictions made by the classification algorithm
    Parameters:
    ----------
    y_true: array of true labels
    y_pred: array of predicted labels
    
    """
    
    
    # Get the unique labels in the sample
    labels = np.unique(y_true)
    
    # Create an empty matrix
    matrix = np.zeros((len(labels), len(labels))) 
    
    # Loop through the data and update the confusion matrix, get true and predicted labels
    for i in range(len(y_true)):
        true_label = y_true[i]
        pred_label = y_pred[i]
        
        matrix[true_label, pred_label] += 1
    
    return matrix


def accuracy_score(y_true, y_pred):
    """
    Returns the accuracy of a classification model
    ----------
    y_true: array of true labels
    y_pred: array of predicted labels
    
    """
    
    # count nb of values where y_true is equal to y_pred, divided by the number of datapoints in the sample
    correct_predictions = np.count_nonzero(y_true == y_pred)
    accuracy_score = correct_predictions / len(y_true)
    return accuracy_score


def calc_precision_recall(y_true, y_pred, calc_confusion_matrix):
    """
    Returns the precision and recall scores, given the confusion matrix generated by calc_confusion_matrix.
    ----------
    y_true: array of true labels
    y_pred: array of predicted labels
    calc_confusion_matrix: matrix generated by calc_confusion_matrix
    """
    # Generate the confusion matrix
    confusion_matrix = calc_confusion_matrix(y_true, y_pred)
    
    # Initialize lists to store the precision and recall values
    precision = []
    recall = []
    
    # Loop through the classes
    for i in range(confusion_matrix.shape[0]):
        # Calculate the precision for class i
        precision_i = confusion_matrix[i, i] / confusion_matrix[:, i].sum()
        precision.append(precision_i)
        
        # Calculate the recall for class i
        recall_i = confusion_matrix[i, i] / confusion_matrix[i, :].sum()
        recall.append(recall_i)
        
    return precision, recall
```

In the cell below, develop your own code for performing 5-fold nested cross-validation along with your implemenation of k-NN above. You must write your own code -- the scikit-learn module may only be used for verification purposes. 

Your code for nested cross-validation should invoke your kNN function (see above). You cross validation function should be invoked similary to:

    accuracies_fold = myNestedCrossVal(X,y,5,list(range(1,11)),['euclidean','manhattan'],mySeed)
    
where X is your data matrix (containing all samples and features for each sample), 5 is the number of folds, y are your known output labels, ``list(range(1,11)`` evaluates the neighbour parameter from 1 to 10, and ``['euclidean','manhattan',...]`` evaluates the distances on the validation sets.  mySeed is simply a random seed to enable us to replicate your results.

**Notes:**

- you should evaluate **at least** two distance functions
- you should evaluate number of neighbours from 1 to 10
- your function should return a list of accuracies per fold
- for each **fold**, your function should print:
  - the accuracy per distinct set of parameters on the validation set
  - the best set of parameters for the fold after validation
  - the confusion matrix per fold (on the testing set)


```python
# parameters for testing code
nFolds = 5
mySeed = 1234567
np.random.seed(mySeed)

# # Creates an array of random permutation of indices between 0 and the length of the X data.
# # The indices are then split up into (folds) folds
# indices = np.random.permutation(np.arange(0,len(X),1))
# indices = np.array_split(indices, nFolds)

k_values = list(range(1,11))
distance_metrics = ['euclidean','manhattan']
```

```python
def myNestedCrossVal(X, y, nFolds, k_values, distance_metrics, mySeed):
    
    # Set the random seed
    np.random.seed(mySeed)
    
    # Creates an array of random permutation of indices between 0 and the length of the X data
    # The indices are then split up into n folds
    indices = np.random.permutation(np.arange(0, len(X), 1))
    indices = np.array_split(indices, nFolds)
    
    # Initialize a list to store accuracies for each fold
    fold_accuracies_list = []
    
    # Loop through the folds
    for fold in range(nFolds):
    # Get the training and test sets for the fold
        X_train = X[indices[fold]]
        y_train = y[indices[fold]]
        X_test = X[~indices[fold]]
        y_test = y[~indices[fold]]

        
        # Initialize lists to store accuracy and confusion matrix for each parameters 
        accuracies_param = []
        confusion_matrices = []
        
        # Loop through the set of parameters (k, distance metric)
        for k in k_values:
            for distance_metric in distance_metrics:
                # Train the k-NN classifier
                y_pred = get_knn(X_train, y_train, X_test, k, distance_metric)
                # Calculate the accuracy
                accuracy = accuracy_score(y_test, y_pred)
                conf_matrix = calc_confusion_matrix(y_test, y_pred)
                # Store the accuracy
                accuracies_param.append(accuracy)
                confusion_matrices.append(conf_matrix)
        
        # Get the index of the set of parameters with the highest accuracy
        best_param_index = np.argmax(accuracies_param)
        
        # Get the best set of parameters
        k_best = k_values[best_param_index // len(distance_metrics)]
        best_distance_metric = distance_metrics[best_param_index % len(distance_metrics)]
        
        
        # Print the accuracy and the best set of parameters and the confusion matrix for the current fold
        print(f"Best accuracy for fold {fold}: {accuracies_param[best_param_index]:.3f}")
        print(f"Best parameters for fold {fold}: k = {k_best}, distance_metric = {best_distance_metric}")
        print(f"Confusion matrix for fold {fold}:\n", confusion_matrices[best_param_index])
        
        # Append the accuracy for the best parameters to the list of accuracies per fold
        fold_accuracies_list.append(accuracies_param[best_param_index])
    return fold_accuracies_list
```

```python
# evaluate clean data code
myNestedCrossVal(X, y, 5, k_values, distance_metrics, mySeed)
```
